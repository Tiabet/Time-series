```python
import matplotlib.pyplot as plt
import numpy as np

# Define the initial conditions and time step
y0 = 0
y1 = 1
dt = 1

# Create empty lists to store the time and y values
t = [0,1]
y = [y0, y1]

# Iterate through the time steps and calculate the y values
for i in range(1, 100):
    y.append(0.8**(1/2)* y[i-1]-0.2*y[i-2])
    t.append(t[i-1] + dt)

# Plot the graph
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('y(t)')
plt.show()
```
