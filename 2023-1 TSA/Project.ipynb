```python
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import kpss


df = pd.read_csv("AirPassengers.csv", names = ['date','passengers'] , index_col=0)
```

```python
#데이터 모양 확인
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(9, 2*3))
axes[0].plot(df,color='black',linewidth=1,label='number of passengers')
axes[0].hlines(xmin=df.index[0],xmax=df.index[-1],y=0,color='gray',linewidth=1)
axes[0].legend()
axes[1].plot(df.diff(),color='black',linewidth=1,label='First-difference of␣number of passengers')
axes[1].hlines(xmin=df.index[0],xmax=df.index[-1],y=0,color='gray',linewidth=1)
axes[1].legend()
plt.show()
```

```python
#데이터 추가 작업
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(9, 2*3))
axes[0].plot(np.log(df),color='black',linewidth=1,label='logged number of passengers')
axes[0].hlines(xmin=df.index[0],xmax=df.index[-1],y=0,color='gray',linewidth=1)
axes[0].legend()
axes[1].plot(np.log(df/df.shift(1)),color='black',linewidth=1,label='First-difference and logged number of passengers')
axes[1].hlines(xmin=df.index[0],xmax=df.index[-1],y=0,color='gray',linewidth=1)
axes[1].legend()
```

```python
#데이터 정상성 검정 - adf test
def adf_test(dataframe):
    result = adfuller(dataframe)
    print(f'Statistics: {result[0]}')
    print(f'p-value: {result[1]}')
    print(f'Critical values: {result[4]}')
adf_test(df)
print('---'*40)
adf_test(df.diff().dropna())
print('---'*40)
adf_test(np.log(df))
print('---'*40)
adf_test(np.log(df/df.shift(1)).dropna())
```

```python
#데이터 정상성 검정 - kpss test
def kpss_test(dataframe):
    result = kpss(dataframe)
    print(f'Statistics: {result[0]}')
    print(f'bounded p-value: {result[1]}')
    print(f'Critical values: {result[3]}')
kpss_test(df)
print('---'*40)
kpss_test(df.diff().dropna())
print('---'*40)
kpss_test(np.log(df))
print('---'*40)
kpss_test(np.log(df/df.shift(1)).dropna())
```

ADF Test와 KPSS Test의 결과, 1차 차분을 진행했을 때와 로그 데이터로 1차 차분을 진행했을 때 데이터가 정상성을 띈다고 판단할 수 있습니다.

```python
df_diff = df.diff().dropna()
df_logged_diff = np.log(df/df.shift(1)).dropna()
```

```python
#1차 차분 데이터 ACF, PACF
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2*4))
plot_acf(df_diff, lags=30, ax=axes[0], title='Autocorrelations',color='black',vlines_kwargs={'colors':'black','linewidth':5}, alpha=None)
plot_pacf(df_diff, lags=30, ax=axes[1], method='ols', title='PACF',color='gray',vlines_kwargs={'colors':'gray','linewidth':5}, alpha=None)
axes[1].hlines(xmin=0,xmax=30,y=2*np.sqrt(1/len(df)),label='two standard deviations from zero',color='black',linewidth=1)
axes[1].hlines(xmin=0,xmax=30,y=-2*np.sqrt(1/len(df)),color='black',linewidth=1)
axes[1].legend()
plt.tight_layout()
plt.show()
```

```python
#로그-차분 데이터 ACF, PACF
f, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2*4))
plot_acf(df_logged_diff, lags=30, ax=axes[0], title='Autocorrelations',color='black',vlines_kwargs={'colors':'black','linewidth':5}, alpha=None)
plot_pacf(df_logged_diff, lags=30, ax=axes[1], method='ols', title='PACF',color='gray',vlines_kwargs={'colors':'gray','linewidth':5}, alpha=None)
axes[0].hlines(xmin=0,xmax=30,y=2*np.sqrt(1/len(df)),label='two standard deviations from zero',color='black',linewidth=1)
axes[0].hlines(xmin=0,xmax=30,y=-2*np.sqrt(1/len(df)),color='black',linewidth=1)
axes[0].legend()
axes[1].hlines(xmin=0,xmax=30,y=2*np.sqrt(1/len(df)),label='two standard deviations from zero',color='black',linewidth=1)
axes[1].hlines(xmin=0,xmax=30,y=-2*np.sqrt(1/len(df)),color='black',linewidth=1)
axes[1].legend()
plt.tight_layout()
plt.show()
```

보였던 것과는 다르게 1차 차분 데이터와 로그-차분 데이터의 ACF, PACF들이 거의 차이가 없지만, 분산을 일정하게 유지하였고 충분히 정상성을 띄는 것으로 확인된 로그-차분 데이터로 진행하겠습니다.

```python
def get_max_value(element):
    if isinstance(element, int) or isinstance(element, float):
        return element
    else:
        return max(element)
```

로그-차분 데이터의 ACF와 PACF를 보고 계수를 선택해보겠습니다. <br>
우선 데이터가 계절성을 가지는 것이 너무나도 명백해보입니다. lag가 12, 24 일 때 ACF, PACF 가 너무 크게 나오기 때문입니다.<br>
따라서 이는 ARMA 모델의 계수로 선정하기엔 적절하지 않아 보여서 제외하겠습니다.<br>
PACF를 보고 AR(p)의 p값으로는 0, 2, 4 정도가 적절해보이고, ACF를 보고 MA(q) 의 q값으로는 0, 1, 4 정도가 적절해보입니다. <br>
더 넓게 보자면 p값으로는 8, 7, [8,7,4], [8,7,4,2,1], [4,2,1] 정도가 있겠고
q값으로는 [8,4,1], [4,1] 정도가 있을 것 같습니다.<br> 적절하게 몇 개만 선택해보겠습니다.

```python
lag_list = [(2, 0, 0), (4,0,0), ([8,4],0,1), ([4,2,1],0,1), (list(range(1,9)),0,1),(list(range(1,8)),0,1), (4,0,[4,1])]
# We will estimate 7 different models.
summary_table = dict()
idx = 0
num_of_obs = len(df_logged_diff[8:])

for lag in lag_list:
    temp_perf_dict = {key: key for key in ['SSE','AIC','SBC','Q(4)','Q(8)','Q(12)']}
    max_values = [get_max_value(elem) for elem in lag]
    max_element = max(max_values)
    res = ARIMA(endog=df_logged_diff[8-max_element:], order=lag, trend='c').fit(method='innovations_mle')
    temp_perf_dict['SSE'] = round(res.sse, 2)
    temp_perf_dict['AIC'] = round(num_of_obs * np.log(res.sse) + 2 * len(res.params), 2)
    temp_perf_dict['SBC'] = round(num_of_obs * np.log(res.sse) + len(res.params) * np.log(num_of_obs), 2)
    q_statistics = res.test_serial_correlation(method='ljungbox', lags=12)[0]
    temp_perf_dict['Q(4)'] = {'q_stats': round(q_statistics[0][3], 2), 'p_val': round(q_statistics[1][3], 2)}
    temp_perf_dict['Q(8)'] = {'q_stats': round(q_statistics[0][7], 2), 'p_val': round(q_statistics[1][7], 2)}
    temp_perf_dict['Q(12)'] = {'q_stats': round(q_statistics[0][11], 2), 'p_val': round(q_statistics[1][11], 2)}
    for param_name, param in zip(res.params.index, res.params):
        temp_perf_dict[param_name] = {'coef': round(param, 2), 't_stats': round(res.tvalues[param_name], 2)}
        hashable_order = tuple([tuple(order) if isinstance(order, list) == True else order for order in res.specification['order']])
        summary_table[hashable_order] = temp_perf_dict
```

```python
table = pd.DataFrame()
for key, value in summary_table.items():
    temp_series = pd.Series(value, name=key)
    table = pd.concat([table,temp_series],axis=1)
    
summary_table.items()
```

```python
# out-of-sample prediction with expanding rolling window.
checker = df_logged_diff.index < '1959-01-01'
train_set = df_logged_diff[checker]
test_set = df_logged_diff[~checker]
n_train = len(train_set)
n_test = len(test_set)
f1, f1_error = [], []
f2, f2_error = [], []
ground_truth = []
for i in range(n_test):
    crt_time = i + n_train
    x_train = df_logged_diff[:crt_time]
    # iterate over the time series data, fitting the model on each expanding window of data
    model_1 = ARIMA(endog = x_train, order=(8,0,1), trend='c').fit(method='innovations_mle') # model 1
    model_2 = ARIMA(endog = x_train, order=(7,0,1), trend='c').fit(method='innovations_mle') # model 2
    # one-step-ahead forecasts
    forecast_1 = model_1.forecast(steps=1)
    forecast_2 = model_2.forecast(steps=1)
    # true one-step-ahead value
    y = df_logged_diff.iloc[crt_time]
    ground_truth.append(y)
    f1.append(forecast_1[0])
    f1_error.append(y-forecast_1[0])
    f2.append(forecast_2[0])
    f2_error.append(y-forecast_2[0])
```

```python
# plot the results
plt.figure(figsize=(12,4))
plt.plot(ground_truth, label='ground truth', color='k', linestyle='--')
plt.plot(f1, label='f1 predicted', color='r')
plt.plot(f2, label='f2 predicted', color='b')
plt.legend()
plt.show()
```
