```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf

# Load the stock price data for Apple Inc.
df = pd.read_csv('https://www.alphavantage.co/query?function=TIME_SERIES_DAILY_ADJUSTED&symbol=AAPL&apikey=R1A10JZ8TS3877SK&datatype=csv&start_date=2010-01-01')

# Plot the ACF of the closing price
plot_pacf(df["close"], title="Apple Stock Price PACF", lags=20)

# Show the plot
plt.show()

import numpy as np

plot_pacf(np.log(df["close"]), title="Logged Apple Stock Price PACF", lags=20)

df['diff'] = df['close'].diff()

df = df.dropna(subset=['diff'])

plot_pacf(df['diff'], title="Differenced Apple Stock Price PACF", lags=20)

plt.show()
```
